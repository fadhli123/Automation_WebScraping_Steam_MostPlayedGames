### Summary Automation System

```python
def update_data():
    # connect to driver
    driver_path = "D:\Installer exe files\Google\chromedriver"
    driver = webdriver.Chrome(executable_path = driver_path)

    # connect to url
    url = "https://store.steampowered.com/charts/mostplayed/"
    driver.get(url)

    # get table
    table = WebDriverWait(driver,15).until(EC.visibility_of_element_located((By.CSS_SELECTOR,".weeklytopsellers_ChartTable_3arZn"))).get_attribute("outerHTML")
    df = pd.read_html(str(table))[0]

    # Get Date & Time
    today = datetime.date.today()
    t = datetime.datetime.now()
    current_time =  t.strftime('%H:%M')

    # Modifying Columns
    df_cleaned = df.drop(columns=['Unnamed: 0']).rename(columns = {'Rank.1':'Game Name', 'Current Players':'Players', 'Peak Today':'Max Players Today'})
    df_cleaned.insert(0, 'Date', today)
    df_cleaned.insert(1, 'Time', current_time)

    # Cleaning Data
    df_cleaned['Game Name'] = df_cleaned['Game Name'].str.replace('not available in your region','').str.replace('(','').str.replace(')','')
    df_cleaned['Game Name'] = df_cleaned['Game Name'].str.replace('\™', '', regex=True).str.replace('\®', '', regex=True).str.replace('\’', '\'', regex=True).str.replace('\雀魂麻將', '', regex=True)

    new_price = df_cleaned['Price'].str.split('Rp', n=3, expand=True)
    new_price[2] = new_price[2].fillna(new_price[1])

    df_cleaned['Price'] = new_price[2]
    df_cleaned['Price'] = df_cleaned['Price'].str.replace(' ','')
    df_cleaned['Price'] = df_cleaned['Price'].fillna(0)
    df_cleaned['Price'] = df_cleaned['Price'].astype('int')

    df_cleaned = df_cleaned.rename(columns={'Price':'Price (Rp)'})

    # Concating Data Frame
    data_frame = pd.read_csv('./Data/Daily Most Played Games on Steam.csv')

    data_frame = data_frame.sort_values(by=['Date', 'Time'], ascending=True)

    output = pd.concat([data_frame,df_cleaned])

    output['Date'] = pd.to_datetime(output['Date'])

    output = output.sort_values(by=['Date', 'Time'], ascending=True)

    output.to_csv('./Data/Daily Most Played Games on Steam.csv', index=False)

    display(output)
        
```

```python
#Set Automation Time to Function
increment = 0
sleep = 0 + increment + increment
while(True):
    update_data()
    time.sleep(1800 - sleep)
    sleep = sleep + 10
    increment = increment + 1
```
